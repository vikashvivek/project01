```python
def CreateQuestionTxt(more_lines,content,title,count):
  f = open('/content/drive/MyDrive/txtdata/vp.txt', 'a')
  if count==0:
    f.writelines('Title: '+title)
  else:
    f.writelines('\n\n\n\nTitle: '+title)
  f.writelines('\n\nContent: '+content)
  jsonData = more_lines
  for x in jsonData:
    q="Question: "+x['question']
    a="Answer: "+x['answer']
    print(q)
    print(a)
    f.writelines('\n\n'+q)
    f.writelines('\n'+a)
  f.close()
```

```python

for i in range(0, df.shape[0]):
  try:
    more_lines=nlp(df['contents'][i])
  except:
    print('error -' ,i)
    pass
  else:
    print('\n',i,'-',more_lines)
    CreateQuestionTxt(more_lines,df['contents'][i],df['title'][i],i)
print(df.shape)
print(i)
CreateQuestionTxt

```
