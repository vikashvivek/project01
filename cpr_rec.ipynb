```r
# Load the library
library(randomForest)
```

```r
data$label<-as.factor(data$label)
table(data$label)
```

```r
sample = sample.split(data,SplitRatio = 0.75) # splits the data in the ratio mentioned in SplitRatio. After splitting marks these rows as logical TRUE and the the remaining are marked as logical FALSE
train1 =subset(data,sample ==TRUE) # creates a training dataset named train1 with rows which are marked as TRUE
test1=subset(data, sample==FALSE)
```

```r
print(train1)
```

```r
rf<-randomForest(label~.,data=train1,ntree = 500,metric = "Accuracy",
importance=TRUE,proximity = TRUE)
table(predict(rf), train1$label)
```

```r
plot(rf)
```

```r
pred = predict(rf, newdata=test1)
table(pred, test1$label)
```

```r
plot(margin(rf, test1$label))
```

```r
CM = table(pred, test1$label)
```

```r
accuracy = (sum(diag(CM)))/sum(CM)
```

```r
print(accuracy)
```
